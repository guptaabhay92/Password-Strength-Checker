#### Converting words in characters

```python
def word_divide_char(input):
    characters=[]
    for i in input:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(tokenizer=word_divide_char)
X=tfidf.fit_transform(X)
```

```python
tfidf.vocabulary_
```

```python
df['password'][0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=.20, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(penalty='l2', multi_class='ovr')
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_train, y_train)
```

#### 0 : Week password
#### 1 : Password strength is good
#### 2 : Password strength is strong

```python
x_predict=np.array(['abc'])
x_predict=tfidf.transform(x_predict)
y_predict=log_reg.predict(x_predict)
y_predict
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
y_predict=xgb.predict(X_test)
y_predict
```
